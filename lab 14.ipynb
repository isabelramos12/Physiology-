```python
# import packages
import statistics
import pandas as pd
import scipy.stats as stats
from scipy.stats import sem
from statistics import mean, stdev
import matplotlib.pyplot as plt
import numpy as np
#include if using a jupyter notebook, remove if using a .py file
%matplotlib inline
```

```python
#Create datasets
f_vc=[2.7,3.7,3.2,3.45,3.2,3.1]
m_vc=[3.3,4.5,3.7]

#display data
f_vc, m_vc
```

```python
#Calculate the means of data
fvc_mean = mean(f_vc)
mvc_mean = mean(m_vc)

#display the mean
fvc_mean, mvc_mean
```

```python
#Calculate the SEM of data
fvc_sem = sem(f_vc)
mvc_sem = sem(m_vc)

#display the SEM
fvc_sem, mvc_sem
```

```python
# build a bar plot
plt.bar(['Female VC', 'Male VC'],[fvc_mean, mvc_mean], yerr=[fvc_sem, mvc_sem], capsize=10)
plt.xlabel('Biological Sex')
plt.ylabel('Vital Capacity (L)')
plt.title('Vital Capacity Differences in Male vs Female')
```
